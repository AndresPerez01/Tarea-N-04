## EJERCICIO 1

```python
import numpy as np

def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def bisection_method(a, b, tol):
    if f(a) * f(b) >= 0:
        print("El método de bisección no puede aplicarse en este intervalo.")
        return None
    
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0

intervals = [(0, 1), (1, 3.2), (3.2, 4)]
tolerance = 0.01

roots = []
for interval in intervals:
    root = bisection_method(interval[0], interval[1], tolerance)
    roots.append(root)
    print(f"La raíz en el intervalo {interval} es aproximadamente: {root}")
```

## EJERCICIO 6

```python
import numpy as np

def volumen_abrevadero(radio, longitud, altura_agua):
  """
  Calcula el volumen de agua en un abrevadero en forma de semicírculo.

  Parámetros:
    radio (float): Radio del semicírculo en metros.
    longitud (float): Longitud del abrevadero en metros.
    altura_agua (float): Altura del agua en el abrevadero en metros.

  Retorno:
    volumen (float): Volumen de agua en el abrevadero en metros cúbicos.
  """
  # Calculamos el área del semicírculo
  area_semicirculo = np.pi * radio ** 2 / 2
  # Calculamos el área del segmento circular
  angulo = 2 * np.arcsin(altura_agua / radio)
  area_segmento = area_semicirculo * (angulo / np.pi)
  # Calculamos el volumen de agua
  volumen = area_segmento * longitud

  return volumen

# Ejemplo de uso
radio = 1  # Radio del semicírculo en metros
longitud = 10  # Longitud del abrevadero en metros
altura_agua = 0.124  # Altura del agua en el abrevadero en metros
volumen = volumen_abrevadero(radio, longitud, altura_agua)
print("Volumen de agua:", volumen, "m^3")
```

```python
import numpy as np

g = 9.81  # m/s^2
k = 0.1  # Ns/m
m = 0.25  # kg
s0 = 300  # m
def s(t):
    return s0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - np.exp(-k * t / m))

def bisection_method(func, a, b, tol):
    if func(a) * func(b) >= 0:
        print("El método de bisección no puede aplicarse en este intervalo.")
        return None
    
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if func(midpoint) == 0:
            return midpoint
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0
a = 0
b = 20
tolerance = 0.01
root = bisection_method(s, a, b, tolerance)
if root is not None:
    print(f"El tiempo que tarda el objeto en golpear el piso es aproximadamente: {root:.2f} segundos")
else:
    print("No se encontró una raíz en el intervalo dado.")
```

```python
def f(x):
    return x**2 - x - 1

def biseccion(a, b, tol):
    n = 0 
    while (b - a) / 2 > tol:
        n += 1
        p = (a + b) / 2
        if f(p) == 0:
            return p, n
        elif f(a) * f(p) < 0:
            b = p
        else:
            a = p
    p = (a + b) / 2
    return p, n

a = 1
b = 2
tol = 10**-4
raiz, iteraciones = biseccion(a, b, tol)
print(f"La raíz aproximada es: {raiz}")
print(f"Número de iteraciones: {iteraciones}")
```
